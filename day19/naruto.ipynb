```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x    = np.linspace(-10,10,100)
cos  = np.cos(x)
cos3 = np.cos(x)*3
sin  = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
fig, ax = plt.subplots(figsize=(17,8))

ax.plot(x, cos, label='CosX', color='r')
ax.plot(x, cos3, label='3 CosX', color='g', linestyle='dashed')
ax.plot(x, sin, label='Sinx-90', color='b', linestyle='dashdot')
ax.plot(x, sin2, label='2 Sinx', color='y', linestyle='', marker='x')


plt.title('Grafik Cos Sin')
plt.xlabel('X-Axis Label')
plt.ylabel('y-Axis Label')
plt.legend(loc='center')

plt.show()
```
